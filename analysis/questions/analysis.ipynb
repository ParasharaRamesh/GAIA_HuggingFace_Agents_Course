```python
import pandas as pd
import json

```

```python
with open('analysis.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df = df.drop(columns=['Level'])
metadata_df = pd.json_normalize(df['Annotator Metadata'])
df = df.drop(columns=['Annotator Metadata'])
df = pd.concat([df, metadata_df], axis=1)
df['Steps'] = df['Steps'].apply(lambda x: x.strip().split('\n') if isinstance(x, str) else [])
df['Tools'] = df['Tools'].apply(lambda x: x.strip().split('\n') if isinstance(x, str) and x.strip() else [])
```

```python
import json

# Load the analysis JSON
with open('analysis.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Extract only task_id and Final answer
filtered_data = [
    {"task_id": item["task_id"], "Final answer": item["Final answer"]}
    for item in data
]

# Save to a new JSON file
with open('task_id_final_answer.json', 'w', encoding='utf-8') as f:
    json.dump(filtered_data, f, indent=4, ensure_ascii=False)

print("✅ task_id_final_answer.json created successfully!")
```
